### Flow Prediction for Wastewater Treatment Plant

### Modeling

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.metrics import mean_squared_error
from datetime import datetime
from sklearn.inspection import permutation_importance
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Ridge
from sklearn import linear_model
```

```python
#load data from Feature Engineering
df = pd.read_csv('data/flow_pred_feature.csv') 
#df.set_index('date',inplace=True)
df.head()
```

### Split Data for Training and Testing

```python
# Split Data for training and testing
X_train  = df[:-100].drop(['wwtp_flow_for','date'], axis = 1)
y_train= df[:-100].wwtp_flow_for
X_test  = df[-100:].drop(['wwtp_flow_for','date'], axis = 1)
y_test= df[-100:].wwtp_flow_for
```

### Setup Base Models for Comparision
Flow Prediction = Mean WWTP Flow for Training Set

```python
# create full array with mean flow value for base model
train_preds = np.full((len(y_train)),y_train.mean())
test_preds = np.full((len(y_test)),y_train.mean())

# Calculation of Mean Aboslute Error (MAE)
b_train_mae = mean_absolute_error(y_train,train_preds)
b_test_mae = mean_absolute_error(y_test,test_preds)

print(f"Base Model Train mae:{b_train_mae:.2f}")
print(f"Base Model Test mae:{b_test_mae:.2f}")
```

### Simple Linear Regression Model

```python
#train and fit regression model
pipe_lg = Pipeline([('scaler', StandardScaler()),
                 ('lg_model', LinearRegression())])
pipe_lg.fit(X_train, y_train)

# Calculation of Mean Aboslute Error (MAE)
lg_train_preds = pipe_lg.predict(X_train)
lg_test_preds = pipe_lg.predict(X_test)
lg_train_mae = mean_absolute_error(y_train,lg_train_preds)
lg_test_mae = mean_absolute_error(y_test,lg_test_preds)

print(f"LG Model Train mae:{lg_train_mae:.2f}")
print(f"LG Model Test mae:{lg_test_mae:.2f}")
```

```python
# Plotting the training flow Predictions
plt.figure(figsize=(12, 4))
plt.plot(np.array(y_train)[:90], label = "Actual Flow at WWTP", marker='o')
plt.plot(lg_train_preds[:90], label = "Predicted Flow at WWTP", marker='o')
plt.legend()
plt.xlabel('Date')
plt.ylabel('Flow (mgd)')
plt.title('Training Set Predictions')
plt.show()
```

```python
# Plotting the test flow Predictions
plt.figure(figsize=(12, 4))
plt.plot(np.array(y_test), label = "Actual Flow at WWTP", marker='o')
plt.plot(lg_test_preds, label = "Predicted Flow at WWTP", marker='o')
plt.legend()
plt.xlabel('Date')
plt.ylabel('Flow (mgd)')
plt.title('Test Set Predictions')
plt.show()
```

```python
# Coefficients used by the Model
lreg_coef = pipe_lg.named_steps['lg_model'].coef_
lreg_fea = X_train.columns
lg_model = pd.DataFrame(data = np.ravel(lreg_coef), index = np.ravel(lreg_fea),columns = [ 'Coef'])
lg_model.sort_values('Coef')
```

```python
perm_importance = permutation_importance(pipe_lg, X_train, y_train, n_repeats=10, random_state=1066)
sorted_idx = perm_importance.importances_mean.argsort()
fig = plt.figure(figsize=(12, 6))
plt.barh(range(len(sorted_idx)), perm_importance.importances_mean[sorted_idx], align='center')
plt.yticks(range(len(sorted_idx)), np.array(X_train.columns)[sorted_idx])
plt.title('Permutation Importance')
plt.show()
```

### Polynomial Regression Model
- The best model was with the poly degree = 1, which indicates that a simple linear regression is a better model
- Results shown for degree for comparison purposes

```python
# Grid Search over polynomial degrees
params = {'poly__degree': [1, 2, 3],
         'poly__include_bias': [True, False]}

pipe_pl = Pipeline([('poly', PolynomialFeatures()),
                 ('scaler', StandardScaler()),
                 ('pl_model', LinearRegression())])


grid = GridSearchCV(pipe_pl, param_grid=params)
grid.fit(X_train, y_train)

print(grid.best_estimator_)
```

```python
# Model with higher oder ploynomials
pipe = Pipeline([('poly', PolynomialFeatures(degree = 2,include_bias = False)),
                 ('scaler', StandardScaler()),
                 ('pl_model', LinearRegression())])
pipe.fit(X_train, y_train)

# Calculation of Mean Aboslute Error (MAE)
pl_train_preds = pipe.predict(X_train)
pl_test_preds = pipe.predict(X_test)
pl_train_mae = mean_absolute_error(y_train,pl_train_preds)
pl_test_mae = mean_absolute_error(y_test,pl_test_preds)

print(f"Poly Model Train mae:{pl_train_mae:.2f}")
print(f"Poly Model Test mae:{pl_test_mae:.2f}")
```

### Ridge Regression
- best results with apha = 10

```python
# Grid Search over Ridge alphas
params = {'ridge__alpha': [.1,1,10,20,40]}

pipe_ridge = Pipeline([('scaler', StandardScaler()),
                       ('ridge', Ridge())])

grid_ridge = GridSearchCV(pipe_ridge, param_grid=params)
grid_ridge.fit(X_train, y_train)
print(grid_ridge.best_estimator_)
```

```python
#train and fit model
ridge = Ridge(alpha=10).fit(X_train,y_train)
  
# Calculation of Mean Aboslute Error (MAE)
ridge_train_preds = ridge.predict(X_train)
ridge_test_preds = ridge.predict(X_test)
ridge_train_mae = mean_absolute_error(y_train,ridge_train_preds)
ridge_test_mae = mean_absolute_error(y_test,ridge_test_preds)

print(f"Ridge Model Train mae:{ridge_train_mae:.2f}")
print(f"RidgeModel Test mae:{ridge_test_mae:.2f}")
```

### Lasso Regression

```python
# Grid Search over Lasso alphas
params = {'lasso__alpha': [.01,.05,.1,.5,1,2]}

pipe_lasso = Pipeline([('scaler', StandardScaler()),
                       ('lasso', linear_model.Lasso())])

grid_lasso = GridSearchCV(pipe_lasso, param_grid=params)
grid_lasso.fit(X_train, y_train)
print(grid_lasso.best_estimator_)
```

```python
#train and fit model
lasso = linear_model.Lasso(alpha=.05).fit(X_train,y_train)
  
# Calculation of Mean Aboslute Error (MAE)
lasso_train_preds = lasso.predict(X_train)
lasso_test_preds = lasso.predict(X_test)
lasso_train_mae = mean_absolute_error(y_train,lasso_train_preds)
lasso_test_mae = mean_absolute_error(y_test,lasso_test_preds)

print(f"Lasso Model Train mae:{lasso_train_mae:.2f}")
print(f"Lasso Model Test mae:{lasso_test_mae:.2f}")
```

### Results
- Models were evaulated on Mean Absolute Error (MAE) to avoid overfitting on the few peak flows that are far from the average. 
- The results of the models used are shown in the table below.
- The Ridge regression model had the best reuslts.
- The most important features were found to be:
    - wwtp flow yesterday
    - wwtp flow today
    - precipitation rain today
    - river flow today
    - precipitation rain forecast
    - temperature minimum yeasterday
