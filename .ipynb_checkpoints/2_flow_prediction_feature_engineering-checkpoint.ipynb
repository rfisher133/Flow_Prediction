### Flow Prediction for Wastwater Treatment Plant

### Feature Engineering

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load data from EDA
df = pd.read_csv('data/flow_pred_data.csv') 
df.set_index('date',inplace=True)
df.head()
```

### Setup New Features that are Forecost (for) and previous day values

Forecast features (new)
- precip_for_am (forecast morning precipitation values)
- precip_for_apm (forecast afternoon precipitation values)
- wwtp_flow_for (TARGET FOR MODEL, forecast wwtp flow)

Todays Feature
- precip_am (morning precipitation)
- precip_amp (afternoon precipitation)
- wwtp_flow (treatment plant flows)
- river_flow (river flows)

Previous day features (new)
 - precip_yes (previous day precipitation)
 - precip_past3d (previous 3 day precipitation)
 - wwtp_flow_yes (previous day wwtp flow)
 - wwtp_flow_past3d (previous 3 day precipitation)
 - river_flow_yes (previous day river flow)

```python
#forecast features
df['precip_for_am'] = df['precip_am'].shift(-1)
df['precip_for_pm'] = df['precip_pm'].shift(-1)
df['wwtp_flow_for'] = df['wwtp_flow'].shift(-1)

#Previous Day Features
df['precip_yes'] = df['precip_daily'].shift(1)
df['wwtp_flow_yes'] = df['wwtp_flow']. shift(1)
df['river_flow_yes'] = df['river_flow'].shift(1)
df['precip_past3d'] = df['precip_daily'].shift(1)+df['precip_daily'].shift(2)+df['precip_daily'].shift(3)
df['wwtp_flow_past3d'] = df['wwtp_flow'].shift(1)+df['wwtp_flow'].shift(2)+df['wwtp_flow'].shift(3)

# Drop daily precip
df.drop('precip_daily', axis= 1, inplace=True)

#drop dates on ends with Nan
df.dropna(inplace=True)
df.head()
```

```python
#Save data to file for next steps
df.to_csv("data/flow_pred_feature.csv", encoding='utf-8')
```
