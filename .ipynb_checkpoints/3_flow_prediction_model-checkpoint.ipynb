### Flow Prediction for Wastewater Treatment Plant

### Modeling

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.metrics import mean_squared_error
from datetime import datetime
from sklearn.inspection import permutation_importance
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Ridge
from sklearn import linear_model
from sklearn.svm import SVR

from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
```

```python
#load data from Feature Engineering
df = pd.read_csv('data/flow_pred_feature.csv') 
#df.set_index('date',inplace=True)
df.head()
```

### Split Data for Training and Testing

```python
# Split Data for training and testing
X=df.drop(['wwtp_flow_for','date'], axis = 1) #drop target and data columns
y=df.wwtp_flow_for
X_train, X_test,y_train,y_test = train_test_split(X,y,test_size=0.3, random_state = 0)
```

### Setup Base Models for Comparision
Base Model Flow Prediction = Mean WWTP Flow for Training Set

```python
# create full array with mean flow value for base model
train_preds = np.full((len(y_train)),y_train.mean())
test_preds = np.full((len(y_test)),y_train.mean())

# Calculation of Mean Aboslute Error (MAE)
b_train_mae = mean_absolute_error(y_train,train_preds)
b_test_mae = mean_absolute_error(y_test,test_preds)
b_test_r2=r2_score(y_test,test_preds)
print(f"Base Model Train mae:{b_train_mae:.2f}")
print(f"Base Model Test mae:{b_test_mae:.2f}")
print(f"Test Score: {b_test_r2:.2f}")
```

### Simple Linear Regression Model

```python
#train and fit regression model
pipe_lg = Pipeline([('scaler', StandardScaler()),
                 ('lg_model', LinearRegression())])
pipe_lg.fit(X_train, y_train)

# Calculation of Mean Aboslute Error (MAE)
lg_train_preds = pipe_lg.predict(X_train)
lg_test_preds = pipe_lg.predict(X_test)
lg_train_mae = mean_absolute_error(y_train,lg_train_preds)
lg_test_mae = mean_absolute_error(y_test,lg_test_preds)
lg_test_r2=r2_score(y_test,lg_test_preds)
print(f"LG Model Train mae:{lg_train_mae:.2f}")
print(f"LG Model Test mae:{lg_test_mae:.2f}")
print(f"Test Score: {lg_test_r2:.2f}")
```

### Polynomial Regression Model
- The best model was with the poly degree = 1, which indicates that a simple linear regression is a better model
- The model overfits with a polynomial degree of 2

```python
# Grid Search over polynomial degrees
params = {'poly__degree': [1, 2, 3],
         'poly__include_bias': [True, False]}

pipe_pl = Pipeline([('poly', PolynomialFeatures()),
                 ('scaler', StandardScaler()),
                 ('pl_model', LinearRegression())])


grid = GridSearchCV(pipe_pl, param_grid=params)
grid.fit(X_train, y_train)

print(grid.best_estimator_)
```

```python
# Model with higher oder ploynomials
pipe_pl = Pipeline([('poly', PolynomialFeatures(degree = 2,include_bias = False)),
                 ('scaler', StandardScaler()),
                 ('pl_model', LinearRegression())])
pipe_pl.fit(X_train, y_train)

# Calculation of Mean Aboslute Error (MAE)
pl_train_preds = pipe_pl.predict(X_train)
pl_test_preds = pipe_pl.predict(X_test)
pl_train_mae = mean_absolute_error(y_train,pl_train_preds)
pl_test_mae = mean_absolute_error(y_test,pl_test_preds)
pl_test_r2=r2_score(y_test,pl_test_preds)
print(f"Poly Model Train mae:{pl_train_mae:.2f}")
print(f"Poly Model Test mae:{pl_test_mae:.2f}")
print(f"Test Score: {pl_test_r2:.2f}")
```

### Ridge Regression

```python
# Grid Search over Ridge alphas
params = {'ridge__alpha': [1,10,20,40,60,80]}

pipe_ridge = Pipeline([('scaler', StandardScaler()),
                       ('ridge', Ridge())])

grid_ridge = GridSearchCV(pipe_ridge, param_grid=params)
grid_ridge.fit(X_train, y_train)
print(grid_ridge.best_estimator_)
```

```python
#train and fit model using best parameters
pipe_ridge = Pipeline([('scaler', StandardScaler()),
                       ('ridge', Ridge(alpha = 60))])

pipe_ridge.fit(X_train,y_train)
ridge_train_preds = pipe_ridge.predict(X_train)
ridge_test_preds = pipe_ridge.predict(X_test)

# scoring
ridge_train_mae = mean_absolute_error(y_train,ridge_train_preds)
ridge_test_mae = mean_absolute_error(y_test,ridge_test_preds)
ridge_test_r2=r2_score(y_test,ridge_test_preds)
print(f"Ridge Model Train mae:{ridge_train_mae:.2f}")
print(f"RidgeModel Test mae:{ridge_test_mae:.2f}")
print(f"RidgeModel Test R2 score:{ridge_test_r2:.2f}")
```

### Lasso Regression

```python
# Grid Search over Lasso alphas
params = {'lasso__alpha': [.001,.01,.05,.1,1,2]}

pipe_lasso = Pipeline([('scaler', StandardScaler()),
                       ('lasso', linear_model.Lasso(max_iter=10000))])

grid_lasso = GridSearchCV(pipe_lasso, param_grid=params)
grid_lasso.fit(X_train, y_train)
print(grid_lasso.best_estimator_)
```

```python
#train and fit model with best parameters
pipe_lasso = Pipeline([('scaler', StandardScaler()),
                       ('lasso', linear_model.Lasso(alpha=.05,max_iter=10000))])
pipe_lasso.fit(X_train,y_train)
lasso_train_preds = pipe_lasso.predict(X_train)
lasso_test_preds = pipe_lasso.predict(X_test)

# Scoring
lasso_train_mae = mean_absolute_error(y_train,lasso_train_preds)
lasso_test_mae = mean_absolute_error(y_test,lasso_test_preds)
lasso_test_r2=r2_score(y_test,lasso_test_preds)
print(f"Lasso Model Train mae:{lasso_train_mae:.2f}")
print(f"Lasso Model Test mae:{lasso_test_mae:.2f}")
print(f"Lasso Model Test R2 Score:{lasso_test_r2:.2f}")
```

### KNeighborsRegressor

```python
# Grid Search 
params = {'knr__n_neighbors': range(1,10),
         'knr__weights':['uniform', 'distance']}

pipe_knr = Pipeline([('scaler', StandardScaler()),
                       ('knr', KNeighborsRegressor())])

grid_knr = GridSearchCV(pipe_knr, param_grid=params)
grid_knr.fit(X_train, y_train)
print(grid_knr.best_estimator_)
```

```python
# KNighborsRegressor with best parameters
pipe_knr = Pipeline([('scaler', StandardScaler()),
                       ('knr', KNeighborsRegressor(n_neighbors = 8, weights = 'distance'))])
pipe_knr.fit(X_train,y_train)
knr_train_preds = pipe_knr.predict(X_train)
knr_test_preds = pipe_knr.predict(X_test)

# Scoring
knr_train_mae = mean_absolute_error(y_train,knr_train_preds)
knr_test_mae = mean_absolute_error(y_test,knr_test_preds)
knr_test_r2=r2_score(y_test,knr_test_preds)
print(f"KNR Model Train mae:{knr_train_mae:.2f}")
print(f"KNR Model Test mae:{knr_test_mae:.2f}")
print(f"KNR Model Test R2 Score:{knr_test_r2:.2f}")
```

### SVR

```python
# Grid Search 
params = {'svr__C': [1,5,10,20,40],
         'svr__epsilon':[0.001,0.01,0.1,1,2]}

pipe_svr = Pipeline([('scaler', StandardScaler()),
                       ('svr', SVR())])

grid_svr = GridSearchCV(pipe_svr, param_grid=params)
grid_svr.fit(X_train, y_train)
print(grid_svr.best_estimator_)
```

```python
# Support Vector Regression
model = SVR(C=20, epsilon=.001)

pipe_svr = Pipeline([('scaler', StandardScaler()),
                       ('svr', model)])
pipe_svr.fit(X_train,y_train)
svr_train_preds = pipe_svr.predict(X_train)
svr_test_preds = pipe_svr.predict(X_test)

# Scoring
svr_train_mae = mean_absolute_error(y_train,svr_train_preds)
svr_test_mae = mean_absolute_error(y_test,svr_test_preds)
svr_test_r2=r2_score(y_test,svr_test_preds)
print(f"SVR Model Train mae:{svr_train_mae:.2f}")
print(f"SVR Model Test mae:{svr_test_mae:.2f}")
print(f"SVR Model Test R2 Score:{svr_test_r2:.2f}")
```

```python
# Dataframe of results
results = pd.DataFrame({'Model':['Base Model', 'Linear Regression', 'Polynomial Regression', 'Ridge Regression', 'Lasso Regression','KNeighborsRegressor','SVR'],
                       'Train MAE':[b_train_mae,lg_train_mae,pl_train_mae,ridge_train_mae,lasso_train_mae, knr_train_mae,svr_train_mae],
                       'Test MAE':[b_test_mae,lg_test_mae,pl_test_mae,ridge_test_mae,lasso_test_mae, knr_test_mae,svr_test_mae],
                       'Test R2 Score':[b_test_r2,lg_test_r2,pl_test_r2,ridge_test_r2,lasso_test_r2, knr_test_r2, svr_test_r2]}).set_index('Model')
results.round(2)
```

### Model Results
- Models were evaulated on Mean Absolute Error (MAE) to avoid overfitting on the few peak flows that are far from the average.
- Ridge and Lasso regression were the best performing models. 
- Polynmial regression overfit on the training data and did not provide good results for this data. 
- The KNeighbors and SVR did well but not as good as the Ridge and Lasso Regression.

### Review of Important Features

```python
# Coefficients used by the Ridge Model
ridge_coef = pipe_ridge.named_steps['ridge'].coef_
ridge_fea = X_train.columns
ridge_model = pd.DataFrame(data = np.ravel(ridge_coef), index = np.ravel(ridge_fea),columns = [ 'Coef'])
ridge_model.sort_values('Coef')
```

### Important Features
The forecast flow at the wastewater treatment plant was influenced by the following main features:
- Recent flow to the treatment plant (wwtp_flow_past3d).
- Precipitation received in the afternoon of the current day (precip_pm).
- precipitation forceast in the morning of the next day (precip_for_am).
- Current day flow at the treatment plant (wwtp_flow). 
- Flow in the adjacent river on the current day (river_flow).
