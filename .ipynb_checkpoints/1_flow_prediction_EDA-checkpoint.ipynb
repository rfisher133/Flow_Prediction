### Flow Predition for Wastewater Treatment Plant

### Data Cleaning and Exploratory Data Analysis



### 1. Data Cleaning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from datetime import datetime
```


### Dataset No.1 : Wastewater Treatment Plant Records
- Data of daily flow and precipitation 

```python
#import flow data from the Wastewater Treatment Plant
wwtp_flow = pd.read_csv('data/wwtp_flow.csv') #records of daily flow from the treatment plant
wwtp_flow['date']=pd.to_datetime(wwtp_flow['date'])
wwtp_flow.set_index('date',inplace=True)
wwtp_flow.drop(columns = ['flow_max','flow_min'],inplace = True) #remove columns no used
wwtp_flow.rename({'precip': 'precip_1'}, axis=1, inplace=True)
wwtp_flow.head()
```

### Dataset No.2: River Flow Data from Nearest USGS Gauging Station
- Data of daily river flow

```python
#import river flow data from USGS file
df_river = pd.read_csv('data/river_flow.csv',parse_dates=['date']) #records of river flow in cfs
df_river= df_river.groupby("date").mean()
df_river.head()
```

### Dataset No. 3:  National Weather Service Monitoring Site
- Data precipitation and min temperature from Westchester County Monitoring Site

```python
#import NWS weather data
df_weather = pd.read_csv('data/temp_precip.csv') 
df_weather['date']=pd.to_datetime(df_weather['date'])
df_weather.set_index('date',inplace=True)
df_weather.rename({'precip': 'precip_2'}, axis=1, inplace=True)
df_weather.head()
```

### Combine data to a single dataframe

```python
#Combine datasets with datetime index
df_full = pd.merge(wwtp_flow, df_river, left_index=True, right_index=True)
df_full = pd.merge(df_full, df_weather, left_index=True, right_index=True)
df_flow = df_full[:579].copy() #Truncate data to 5/1/2021 to 11/29/2022 where WWTP flow data is erratic
df_flow.head()
```

```python
df_flow.info()
```

#### Fill in missing values

```python
#replace missing wwtp precip with NWS precip
df_flow = df_flow.fillna({'precip_1': df_flow['precip_2']})
df_flow.rename(columns = {'precip_1': 'precip','flow_avg':'wwtp_flow'}, inplace=True)
df_flow.drop(columns = 'precip_2', inplace = True) #drop precip from NWS after filling in missing wwtp precip
#set missing vlues to mean temp
df_flow['tmin'] = df_flow['tmin'].replace(np.nan, df_flow['tmin'].mean()) 
df_flow.info()
```

```python
#Save data to file for next steps
df_flow.to_csv("data/flow_pred_data.csv", encoding='utf-8')           
```

### 2. Exploratory Data Analysis

#### Graph of flow to the treatment plant
- flow spikes due to preciptation can be seen
- Some flow spikes exceed the capacity of the treatment plant (red line)

```python
# Plotting the wwtp flow records
plt.figure(figsize=(12, 4))
plt.plot(df_flow['wwtp_flow'], label = "Daily Flow at WWTP", marker='o')
plt.plot(df_flow['precip'], label = "Precipitation (inches)")
plt.axhline(y=29, color = 'Red', label = "Treatment Capacity")
plt.legend()
plt.xlabel('Date')
plt.ylabel('Flow (mgd)')
plt.title('Flow at Treatment Plant')
plt.show()
```

#### Graph of river flows 
- flow increases from precipitation events similar to the treatment plant

```python
# Plotting the river flow records
plt.figure(figsize=(12, 3))
plt.plot(df_flow['river_flow'], label = "River Flow", marker='o')
plt.legend()
plt.xlabel('Date')
plt.ylabel('Flow (cfs)')
plt.title('River Flow')
plt.show()
```

#### Graph of temperature from NWS monitoring site
- precipitation is measure in liquid (inches) and could be as rain or snow
- freezing line in red denotes when the precipration could be snow instead of rain

```python
# Plotting the weather records
plt.figure(figsize=(12, 4))
plt.plot(df_flow['tmin'], label = "Min Temp", marker='o')
plt.axhline(y=32, color = 'Red', label = "Freezing Line (32F)")
plt.legend()
plt.xlabel('Date')
plt.ylabel('Temp (F)')
plt.title('Weather Data')
plt.show()
```

#### Reviewing correlations of features

```python
df_corr = df_flow.corr()
plt.figure(figsize = (16,5))
sns.heatmap(data  = df_corr, annot = True)
```
